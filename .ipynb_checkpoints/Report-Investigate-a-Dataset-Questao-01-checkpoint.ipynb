```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

% matplotlib inline

df = pd.read_csv('movie-data.csv')
```

```python
df.hist(figsize=[12, 12]);
```

```python
# Obtem lista de gêneros
lista = []
for g in df.genres.unique():
    lista.extend(  g.split('|') ) 
generos = set(lista)

dramas =  df[df.genres.str.contains('Drama')]
```

```python
year = df.release_year.unique()
year.sort()
```

```python
movie_groups = []

for g in generos:    
    # Cria um DataFrame para o gênero g com as colunas popularity, release_year and genres
    movie_gen = df[df.genres.str.contains(g)][['popularity', 'release_year', 'genres']]
    # Configura a coluna gênero para cada registro
    movie_gen['genres_1'] = np.repeat(g, movie_gen.shape[0])
    
    movie_groups.append(movie_gen)

# Concatena todos os gêneros de filmes
movies = pd.concat(movie_groups)
```

```python
movies.describe()
```

```python
period = 5 # 5 Anos

# Cria uma coluna com o Ano ajustado de 5 em 5 anos.
movies['quiquenio'] = (movies.release_year // period)*period
```

```python
# Tabela com a Média de Popularidade no período de 5 em 5 anos, por Gênero de filme
pivot_table = pd.pivot_table(movies, values='popularity', index=['quiquenio'], columns=['genres_1'], aggfunc=np.mean)
pivot_table
```

```python
pivot_table.info()
```

```python
# Os gêneros Foreign, TV Movie e Western possuem dados imcompletos e serão ignorados. 
del pivot_table[['Foreign', 'TV Movie', 'Western']]
```

```python
sns.set()
pivot_table.plot(kind='bar',subplots=True,figsize=(10,35))
```
