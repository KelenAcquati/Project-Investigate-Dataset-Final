```python
#df.groupby('genres')['release_year'].mean().count().plot(kind='bar', title='Average genres by year')
#df.groupby(['release_year','genres'])['popularity'].mean()

# Obtem lista de gÃªneros
lista = []
for g in df.genres.unique():
    lista.extend(  g.split('|') ) 
generos = set(lista)

dramas =  df[df.genres.str.contains('Drama')]
dramas.head()
```

```python
generos
```

```python
year = df.release_year.unique()
year.sort()
```

```python
movie_groups = []
for g in generos:    
    movie_gen = df[df.genres.str.contains(g)][['popularity', 'release_year', 'genres']]
    movie_gen['genres_1'] = np.repeat(g, movie_gen.shape[0])
    movie_groups.append(movie_gen)

movies = pd.concat(movie_groups)
    
   #movies.groupby('release_year', as_index=False).popularity.sum()       

    

    

    
```

```python
#movies.groupby(['release_year', 'genres_1']).popularity.sum()
movies.describe()
```

```python
period = 5
movies['quiquenio'] = (movies.release_year // period)*period
quiq = movies.groupby(['quiquenio', 'genres_1'], as_index=False)
#movies.groupby([ 'release_year','genres_1'],as_index=True).popularity.mean()
#movies.release_year.unique().size
quiq.mean().head(4)
#quiq.popularity.mean().plot(kind="bar",figsize=(20,10))
```

```python
mean_pop = quiq.mean()
anos = mean_pop.quiquenio
#f, ax = plt.subplots(figsize=(15, 10))
#sns.set_context("poster")

# FAIL sns.barplot(x='quiquenio', y='popularity', hue='genres_1',data=mean_pop)
movies.head(3)

pd.pivot_table(movies, values='popularity', index=['quiquenio'], columns=['genres_1'], aggfunc=np.mean)
```
