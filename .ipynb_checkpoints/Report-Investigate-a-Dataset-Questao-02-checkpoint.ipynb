```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

% matplotlib inline

sns.set()
df = pd.read_csv('movie-data.csv')
```

```python
df.hist(figsize=(15,15));
```

```python
df.info()
```

```python
df.describe()
```

```python
# Remover registros com revenue zero

high_revenue = df[df.revenue >= 1e+9]
high_revenue.describe()
```

```python
colunas = ['popularity','budget','revenue','runtime','release_year','genres','director']
high_revenue[colunas].sort_values('revenue',ascending=False)
```

### Questão 3
A receita  e o orçamento de um filme podem qualificar uma produtora de filmes?

```python
colunas = ['popularity', 'budget','revenue','production_companies', 'director', 'release_year']
h = df[df.revenue > 8e+8]
h[colunas].groupby(['director','revenue']).mean()

#h[h.director.str.contains("\|")][['director','revenue']]

```
